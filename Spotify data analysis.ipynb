```python
# set the index to the release_date column
```

```python
tracks.set_index("release_date",inplace=True)
tracks.index=pd.to_datetime(tracks.index)
tracks.head()
```

```python
# Converting duration to miliseconds into seconds
```

```python
tracks["duration"]=tracks["duration_ms"].apply(lambda x:round(x/1000))
tracks.drop("duration_ms",axis=1,inplace=True)
```

```python
tracks
```

```python
# Visualisation with correlation Map
```

```python
corr_df=tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,16))
heatmap=sns.heatmap(corr_df,annot=True)
heatmap.set_title("corrlations Heatmap between variable")
```

```python
# taking 40% of data for sample data
```

```python
sample=tracks.sample(int(0.004*len(tracks)))
```

```python
# printing the len of this data
```

```python
print(len(sample))
```

```python
# Regression plot between loudness and energy
```

```python
plt.figure(figsize=(12,6))
sns.regplot(data=sample,y="loudness",x="energy",color="c").set(title="Loudness vs energy correlation")
```

```python
# Regression plot between popularity and acousticness
```

```python
plt.figure(figsize=(12,6))
sns.regplot(data=sample,y="popularity",x="acousticness",color="b").set(title="popularity vs acousticness correlation")
```

```python
# creating a new column "year" from release_date columns
```

```python
tracks["dates"]=tracks.index.get_level_values('release_date')
tracks.dates=pd.to_datetime(tracks.dates)
years=tracks.dates.dt.year
```

```python
# Distribution plot to visualise the total number of songs in each year since 1922 those are available on spotify music app
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="Number of songs per year")
```

```python
# Duration of songs over years
```

```python
total_dr=tracks.duration
fig_dims=(30,15)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years,y=total_dr,ax=ax,errwidth=False).set(title="year vs duration")
plt.xticks(rotation=90)
plt.show()
```

```python
# Line plot to show the average duration of the songs over the years
```

```python
total_dr=tracks.duration
sns.set_style(style="whitegrid")
fig_dims=(10,5)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years,y=total_dr,ax=ax).set(title="Year vs Duration")
plt.xticks(rotation=60)
plt.show()
```

```python
# Load the spotify features dataset
```

```python
features=pd.read_csv(r"C:\Users\ravi kumar\Downloads\archive (2)\SpotifyFeatures.csv")
features.head()
```

```python
# The duration of songs for different Genres
```

```python
plt.title("Duration of the songs in different Genres")
sns.color_palette("rocket",as_cmap=True)
sns.barplot(y="genre",x="duration_ms",data=features)
plt.xlabel("Duration in miliseconds")
plt.ylabel("Genres")
plt.figure(figsize=(20,10))
```

```python
# Top 5 Genres by popularity
```

```python
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,5))
famous=features.sort_values("popularity",ascending=False).head(10)
sns.barplot(y="genre",x="popularity",data=famous).set(title="Top 5 genres by popularity")
```
